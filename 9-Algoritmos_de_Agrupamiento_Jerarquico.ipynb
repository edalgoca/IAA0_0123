## Realizamos las importaciones necesarias 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score
```

## Cargamos el dataset de iris

```python
iris = load_iris()
X = iris.data
y = iris.target
X
```

## Calculamos el numero optimo de cluster usando silhuette_scores

```python
silhouette_scores = []
for n_clusters in range(2, 11):
    clustering = AgglomerativeClustering(n_clusters=n_clusters)
    clustering.fit(X)
    silhouette_scores.append(silhouette_score(X, clustering.labels_))

optimal_n_clusters = np.argmax(silhouette_scores) + 2
print("Numero oprtimo de cluster es:", optimal_n_clusters)
```

## Mejorar el funcionamiento del custer jerarquico con el numero optimo calculado

```python
clustering = AgglomerativeClustering(n_clusters=optimal_n_clusters)
clustering.fit(X)
```

## Generacion de matriz

```python
Z = linkage(X, method='ward')
```

## Realizamos el plot del grafico dendrogram

```python
plt.figure(figsize=(16, 6))
plt.title("Dendrogram")
dendrogram(Z, labels=y)
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.tight_layout()
plt.show()
```

## Realizamos la grafica con la informacion del cluster jerarquico

```python
plt.scatter(X[:, 0], X[:, 1], c=clustering.labels_, cmap='viridis')
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("Hierarchical Clustering")
plt.show()
```
