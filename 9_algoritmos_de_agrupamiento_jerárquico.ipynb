En que consisten los algoritmos de agrupamiento - jerárquico: Son técnicas utilizadas en el campo de la minería de datos y el aprendizaje automático para agrupar objetos similares en grupos o clústeres. A diferencia de los algoritmos de agrupamiento no jerárquicos, que determinan directamente el número de clústeres, los algoritmos de agrupamiento jerárquico construyen una estructura de clústeres en forma de árbol o dendrograma.

La idea principal detrás de los algoritmos de agrupamiento jerárquico es agrupar objetos similares en clústeres y luego agrupar clústeres similares en clústeres más grandes. Este proceso continúa hasta que todos los objetos estén agrupados en un solo clúster o hasta que se cumpla un criterio de detención definido previamente. Los enfoques principales en los algoritmos de agrupamiento jerárquico son aglomerativo y divisivo.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram

iris = load_iris()
data = iris.data

Z = linkage(data, method='ward')

plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.title('Dendrograma')
plt.xlabel('Índice de muestra')
plt.ylabel('Distancia')
plt.show()
```


Explicación de resultados: El código realiza un agrupamiento jerárquico utilizando el método de enlace ward en el conjunto de datos Iris. Luego genera y muestra el dendrograma que representa la estructura jerárquica de los clústeres obtenidos. El dendrograma proporciona una visualización de cómo se agrupan los datos en base a su similitud. El dendrograma es una representación gráfica que ayuda a comprender la estructura jerárquica de los clústeres y permite identificar diferentes niveles de agrupamiento. Puede ayudar a determinar el número óptimo de clústeres al observar las longitudes de las uniones y la coherencia de los grupos resultantes.
