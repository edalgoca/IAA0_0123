```python
# Importamos las librerias necesarias
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

**Mostramos el dataset.**

```python
data = load_breast_cancer()
dataset = pd.DataFrame(data=data['data'], columns=data['feature_names'])
dataset
```

**Realizamos la importacion para entrenar y testear nuestro model y agragamos un 33% de data para realizar el test, tambien hacemos una separacion de la data**

```python
from sklearn.model_selection import train_test_split
X = dataset.copy()
y = data['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(ccp_alpha=0.01)
clf = clf.fit(X_train, y_train)
```

```python
clf.get_params()
```

```python
X_test
```

## Ahora Despues de visualizar nuestro test vamos a predecir si es maligno benigno donde 0 es Maligno y 1 es Benigno ##

```python
predictions = clf.predict(X_test)
predictions
```

```python
clf.predict_proba(X_test)
```

## Calculamos el score de nuestro algoritmo teniendo en cuenta que tiene que estar cerca de 1 ##

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions, labels=[0,1])
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, predictions)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions, target_names=['malignant', 'benign']))
```
