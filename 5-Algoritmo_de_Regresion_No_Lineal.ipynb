```python
##Importamos la libreria
import seaborn as sns
```

```python
##Utilizo el dataframe anscombe de colab
df = sns.load_dataset("anscombe")
df 
```

```python

sns.set_style("darkgrid")
#Utilizamo el modulo LMPLOT  y lo filtramos con un modelo de orden superior
sns.lmplot (x="x", y="y", data= df, order=2,line_kws={"color":"red"})
```

```python
from pandas.core.arrays.categorical import Ordered
#Realizamos el filtrado de los datos donde la columna data set contenga el valor II 
df_filter = df [ df["dataset"]=="II"]
df_filter.sort_values(by=['x'])
```

```python

sns.set_style("darkgrid")

#Utilizamo el modulo LMPLOT  y lo filtramos con un modelo de orden superior
sns.lmplot (x="x", y="y", data= df_filter, order=2,line_kws={"color":"red"})
```

Podemos ver que al inicio tanto el valor de Y y Y a medida subi el valor de X tambien sube el valor de Y, tienen el mismo comportamiento; luego llego a un punto donde X aumenta y Y desminuye.

Vimos que con el modulo **Seaborn** (SN) con el metodo **LMPLOT** y filtrando de un conjunto de datos II el dataframe y podemos obtener el algoritmo y por medio de **line_kws ** podemos definir la linea roja.

En terminos generales podemos decir que la regresión no lineal es un método para encontrar un modelo no lineal para la relación entre la variable X y un conjunto de variables independientes. A diferencia de la regresión lineal, que está restringida a la estimación de modelos lineales, la regresión no lineal puede estimar modelos con relaciones arbitrarias entre las variables independientes y las dependientes.
