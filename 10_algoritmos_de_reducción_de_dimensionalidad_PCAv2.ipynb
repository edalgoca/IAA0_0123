**10 - algoritmos de reducción de dimensionalidad - PCA**




**En que Consiste el Algoritmo:** 
En reducir la dimensionalidad de los datos, facilitando la visualización, la interpretación y el análisis de datos, y puede ser utilizado en diversas aplicaciones como la exploración de datos, la eliminación de ruido, la compresión de datos y la mejora de la eficiencia computacional en modelos de aprendizaje automático.


**Explicación de resultados:**  
- El objetivo es reducir la dimensionalidad de las características a 2 componentes principales utilizando el algoritmo PCA.

- Después de cargar el dataset, de utiliza el pc2 para obtener solo los 2 componentes principales más importantes. Posteriormente de transforman los datos originales a su proyección en el nuevo espacio de características utilizando fit_transform.

- Como ultimo punto, utilizando los dos componentes principales como ejes X e Y, se colorean según la clase de iris correspondiente para visualizar cómo las muestras se agrupan en función de las componentes principales.



**DataSet:** iris = load_iris()


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Cargar el dataset de ejemplo (Iris)
iris = load_iris()
X = iris.data
y = iris.target

# Aplicar PCA para reducir la dimensionalidad a 2 componentes
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Graficar los puntos proyectados en el nuevo espacio de características
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.title("PCA - Reducción de dimensionalidad")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.show()

```
