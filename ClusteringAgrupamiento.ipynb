```python
#Se importan librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
#se cargan datos
datos = np.loadtxt("sample_data/casas.csv", delimiter=",")
#los años de construcción de una casa contra el precio de la misma
#Se identifican los cluster con una cercanía de dos con un número mínimo de 10 para formar el cluster
clusters = DBSCAN(eps=2, min_samples=10).fit_predict(datos)
print(clusters)

```

```python
#Se gráfica em matplotlib para mostrar los Clusters
plt.figure(figsize=(7.5, 7.5))
plt.scatter(datos[:, 0], datos[:, 1], c=clusters, s=100)
plt.xlabel("Antigüedad de la Casa")
plt.ylabel("Precio de Casa")
plt.box(False)
plt.show()
```

```python
#El más optimo encontrado fue con una cercanía esp=2
#y un mínimo de 10 para formar el cluster
```
