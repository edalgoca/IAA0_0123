```python
#Se importanlas librerias
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

#Cargando los datos desde el archivo CSV
data = pd.read_csv('sample_data/Jerarquico.csv')

```

```python
#se extraen las características relevantes para el agrupamiento
X = data[['Example1', 'Example2', 'Example3']]  

#se Calcula la matriz de distancias
dist_matrix = pdist(X)

#se realiza el agrupamiento jerárquico
linkage_matrix = linkage(dist_matrix, method='ward')
labels = AgglomerativeClustering(n_clusters=3, linkage='ward').fit_predict(X)

```

```python
#Se gráfica el Dendrograma
dendrogram(linkage_matrix)
plt.xlabel('Índices de datos')
plt.ylabel('Distancia')
plt.title('Dendrograma')
plt.show()
```

```python

#Se Imprimen los resultados del agrupamiento
print("Etiquetas de los clústeres:", labels)
```
