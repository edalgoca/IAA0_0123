```python
#se importan las librerias
import cv2
import matplotlib.pyplot as plt

#se cargan las imagen en escala de grises (bajado de internet)
image = cv2.imread('sample_data/pexels-photo-1526403.jpeg', cv2.IMREAD_GRAYSCALE)

```

```python
#Se aplica el filtro Gaussiano para suavizar la imagen
smoothed = cv2.GaussianBlur(image, (5, 5), 0)

#Se aplica el algoritmo de Canny para detectar bordes
edges = cv2.Canny(smoothed, 100, 200)

```

```python
#Se Muestra la imagen original, la imagen suavizada y los bordes detectados
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagen original')

plt.subplot(1, 3, 2)
plt.imshow(smoothed, cmap='gray')
plt.title('Suavizado (Filtro Gaussiano)')

plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Bordes (Algoritmo de Canny)')

plt.tight_layout()
plt.show()
```
