**9 - algoritmos de agrupamiento - jerárquico**




**En que Consiste el Algoritmo:** 
En construir una estructura jerarquica de clusters en el que los objetos o puntos de datos se agrupan en función de su similitud, y los clusters se fusionan iterativamente hasta que se obtiene una estructura jerarquica completa.


**Explicación de resultados:**  

- En este ejemplo, utilizamos la función make_blobs de scikit-learn para generar un conjunto de datos con tres clusters.

- Luego, aplicamos el algoritmo de agrupamiento jerárquico (AgglomerativeClustering) sin un número específico de clusters (n_clusters=None) y con un umbral de distancia de corte igual a cero (distance_threshold=0).

- Después de ajustar el modelo de agrupamiento jerárquico, obtenemos las etiquetas de los clusters (labels) para cada punto de datos. 

- Luego, graficamos los puntos de datos coloreados según los clusters a los que pertenecen.

- El dendrograma muestra la estructura jerárquica de los clusters y las distancias entre ellos.

**DataSet:** fetch_20newsgroups


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram

# Generar datos de ejemplo
X, y = make_blobs(n_samples=50, centers=3, random_state=42)

# Aplicar el algoritmo de agrupamiento jerárquico
clustering = AgglomerativeClustering(n_clusters=None, distance_threshold=0)
clustering.fit(X)

# Obtener las etiquetas de los clusters
labels = clustering.labels_

# Graficar los puntos y los clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("Agrupamiento jerárquico")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()

# Generar el dendrograma
distances = clustering.distances_
linkage_matrix = np.column_stack([clustering.children_, distances, np.zeros_like(distances)]).astype(float)
dendrogram(linkage_matrix)
plt.title("Dendrograma")
plt.xlabel("Índice del punto")
plt.ylabel("Distancia")
plt.show()
```
