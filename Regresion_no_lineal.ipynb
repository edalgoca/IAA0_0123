**Algoritmos de regresión no lineal**

Los algoritmos de regresión no lineal son métodos que permiten modelar relaciones no lineales entre variables. Utilizan funciones no lineales y técnicas de optimización para ajustar los datos y encontrar la mejor curva o superficie que se ajuste a los puntos de datos. Estos algoritmos son útiles cuando los datos exhiben patrones no lineales y permiten realizar predicciones más precisas en diversas áreas de aplicación.

**Codigo Fuente**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

np.random.seed(0)
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = 0.5 * X**3 - X**2 + 2 * X + np.random.normal(0, 1, size=(100, 1))

poly_features = PolynomialFeatures(degree=3)
X_poly = poly_features.fit_transform(X)

model = LinearRegression()

model.fit(X_poly, y)

y_pred = model.predict(X_poly)

# Calcular el error cuadrático medio y el coeficiente de determinación
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Mean Squared Error:", mse)
print("Coefficient of determination:", r2)

plt.scatter(X, y, s=10, label='Datos reales')
plt.plot(X, y_pred, color='red', label='Regresión no lineal')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

**Explicación de resultados de la implementación del algoritmo**

Este algoritmo implementa la regresión no lineal utilizando la regresión polinómica. Genera un conjunto de datos, crea características polinómicas a partir de los datos de entrada, ajusta un modelo de regresión lineal a los datos transformados y realiza predicciones. Luego, calcula y muestra métricas de evaluación y visualiza los datos reales y la línea de regresión en un gráfico que representa los datos reales como puntos dispersos y la línea de regresión no lineal ajustada utilizando el color rojo, tambien utiliza la clase para crear características polinómicas a partir de los datos de entrada X.
