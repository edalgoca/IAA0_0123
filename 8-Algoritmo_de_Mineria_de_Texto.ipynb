```python
# Impotamos funcio NLT que proporciona funciones que pueden tokenizar un texto en palabras u oraciones
import nltk
```

```python
#Tokenize nos sirve para provee un analizador léxico y convirtiendo la entrada en tokens
from nltk.tokenize import sent_tokenize
text = '''Make a career of humanity. Commit yourself to the noble struggle for equal rights. You will make a better person of yourself, a greater nation of your country, and a finer world to live in.'''
tokenized_text = sent_tokenize(text)
print(tokenized_text)
```

```python
## Se dividen cadenas en listas de subcadenas.
from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(text)
print(tokenized_word)
```

```python
from nltk.probability import FreqDist
frequency = FreqDist(tokenized_word)
print(frequency)
```

```python
#Enumere los n elementos más comunes y sus recuentos desde el más común hasta el menos común y lo realizmo por top 10
frequency.most_common(10)
```

```python
## Graficamos los 10 text mas comunes
import matplotlib.pyplot as plt
frequency.plot(10, cumulative=False)
plt.show()
```

```python
from nltk.corpus import stopwords
## Verificamos las Stop Words o palabras vacía a la que se hace referencia a palabras que no están en bots o carecen de sentido
stop_words=set(stopwords.words("english"))
print(stop_words)
```

```python
print(len(stop_words))
```

```python
#Eliminamos los stop word en la variable new text
text = '''Make a career of humanity. Commit yourself to the noble struggle for equal rights. You will make a better person of yourself, a greater nation of your country, and a finer world to live in. career'''
text_original = text
words = [word for word in text.split() if word.lower() not in stop_words]
new_text = " ".join(words)
print("------Texto Original-------------------------------")
print(text_original)
print("------Texto nuevo si los stopwords-------------------------------")
print(new_text)

print("Old length: ", len(text))
print("New length: ", len(new_text))
```

```python
## Se dividen cadenas en listas de subcadenas.
from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(new_text)
print(tokenized_word)

from nltk.probability import FreqDist
frequency = FreqDist(tokenized_word)
print(frequency)

##Hacemos la grafica ya con los text mas reducidos sin palabras stopword.
import matplotlib.pyplot as plt
frequency.plot(10, cumulative=False)
plt.show()
```


En el primer grafico vemos que las palabras que mas aparecen y que no nos dan mucho sentido, luego de aplicar funcion **stop_words** realizamos la limpieza de nuestro texto original.
